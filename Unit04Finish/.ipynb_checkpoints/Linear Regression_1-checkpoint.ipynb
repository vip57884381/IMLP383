# Linear Regression


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## 1. 線性迴歸基本概念

#### 直線

$$f(x) = 1.2x + 0.8$$



```python
#準備 50 個點
#5-0/50 = 0.1 +....
# 0~5線性切割
x = np.linspace(0, 5, 50)
print(x)
```

```python
y = 1.2*x + 0.8
y
```

```python
#畫圖
plt.scatter(x,y)
plt.plot(x, 1.2*x+0.8, 'r')
```

#### add noise 

$$f(x) + \varepsilon(x)$$


```python
#加雜訊
y = 1.2*x + 0.8 + 0.6*np.random.randn(50)
print(y)
```

```python
#在劃一次
plt.scatter(x,y)
plt.plot(x, 1.2*x + 0.8, 'r')
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
```

```python
regr
```

要求資料轉換<br>

$$[x_1, x_2, \ldots, x_{50}]$$

==> 

$$[[x_1], [x_2], \ldots, [x_{50}]]$$


矩陣的概念

```python
#確認 X 資料維度
print(x.shape)      # 1維
X = x.reshape(-1,1)
X.shape
```

```python
regr.fit(X,y)
```

```python
#predict預測 = 機器學習 學習出來的答案
Y = regr.predict(X)
print(Y)
```

```python
#了解你的訓練參數
regr.get_params()
```

```python
plt.scatter(x, y)               #原本 
plt.plot(x, Y, 'ro')            #預測
#calculate mse
mse = np.sum((Y-y)**2) / len(y) #加總((原本-預測)^2) / 總樣本數
print(mse)
```

```python
#Y=WX+b
W=regr.coef_#斜率
W

```

```python
b=regr.intercept_ #截距
b
```

```python
Y
```

```python
#可以將新的x資料帶入算出結果
W*3+b
```
